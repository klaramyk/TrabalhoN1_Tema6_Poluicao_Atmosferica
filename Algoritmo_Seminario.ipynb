```python
import csv #https://docs.python.org/3/library/csv.html
#importar biblioteca que vai ler o arquivo CSV

import urllib.request #https://docs.python.org/3/library/urllib.request.html#module-urllib.request
#importar biblioteca que vai ler o link do github com o arquivo CSV

url_arquivo_dados = 'https://raw.githubusercontent.com/klaramyk/TrabalhoN1_Tema6_Poluicao_Atmosferica/main/suape_janeiro_2022.csv'

with urllib.request.urlopen(url_arquivo_dados) as leitor:
  linhas = leitor.read().decode('utf-8').splitlines()

dados = [] #definindo essa variável como lista vazia

for linha in csv.reader(linhas):
  dados.append(linha) #adiciona linhas do arquivo na lista, sempre ao final

for linha in dados [:10]: #mostra as primeiras 10 linhas da lista, começando da posição 0 e indo até a 9
  print(linha)
```

```python
from tabulate import tabulate
#documentação: https://pypi.org/project/tabulate/
#para formatação (design de tabela)

table = []

cabecalho_completo = dados.pop(0) #.pop: tira o cabecalho e coloca nesta variável
cabecalho = [cabecalho_completo[0], cabecalho_completo[4],
             cabecalho_completo[5], cabecalho_completo[6]]

for cada_linha in dados [0:]: #definindo uma nova variável dentro da lista inicial, pegando desde o 1° termo (data)
  valores_info = [info.strip() for info in cada_linha]

  #variável para inserir as infos necessárias para o cálculo de média diária
  #dentro dos [] está a posição na linha, começando em 0 (data)
  data = valores_info[0]
  poluente = valores_info[4]
  valor = float(valores_info[5])
  unidade = valores_info[6]

  table.append([data, poluente, valor, unidade])

print(tabulate(table, headers = cabecalho, tablefmt = 'heavy_outline'))
```
