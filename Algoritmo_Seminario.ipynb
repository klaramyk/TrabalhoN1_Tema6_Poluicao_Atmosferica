```python
import csv #https://docs.python.org/3/library/csv.html
#importa biblioteca que vai ler o arquivo CSV

import urllib.request #https://docs.python.org/3/library/urllib.request.html#module-urllib.request
#importa biblioteca que vai ler o link do github com o arquivo CSV

url_arquivo_dados = 'https://raw.githubusercontent.com/klaramyk/TrabalhoN1_Tema6_Poluicao_Atmosferica/main/suape_janeiro_2022.csv'

with urllib.request.urlopen(url_arquivo_dados) as leitor:
  linhas = leitor.read().decode('utf-8').splitlines()
  dados = {} #dict para armazenar os valores do poluente data:valor
  unidades = {} #dict para armazenar a unidade de medida

  #itera sobre o arquivo para definir a posição dos dados necessários
  for linha in csv.reader(linhas[1:]):
    data = linha[0]
    poluente = linha[4]
    valor = float(linha[5])
    unidade = linha[6]

    unidades[poluente] = unidade

    if poluente in dados:

      if data in dados[poluente]:
        dados[poluente][data].append(valor)
        #adiciona o valor no final da lista já existente

      else:
        dados[poluente][data] = [valor]
        #adiciona valor na lista de poluente, caso não exista ainda

    else:
      dados[poluente] = {data: [valor]}
      #adiciona novo poluente = chave:valor na lista

opcoes = list(dados.keys())
#exibe todas as chaves do dict (nomes dos poluentes) como um índice
#https://stackoverflow.com/questions/16819222/how-do-i-return-dictionary-keys-as-a-list-in-python

print('***** Dados de Emissão de Poluentes Atmosféricos ***** \n')
print('► Estação de Coleta: Suape/PE')
print('► Período de Referência: Janeiro/2022 \n')
print('Índice de Poluentes')

indice = 1

for opcao in opcoes:
  print(f'{indice}) {opcao}')
  indice = indice + 1
print(f'{indice}) Todos os poluentes \n')

escolha = input('Para qual número você quer ver a média diária? ')

print('\n--------------------------------------')

#https://docs.python.org/3/library/stdtypes.html#str.isdigit
if escolha.isdigit(): #verifica se a string é um número
  escolha = int(escolha)

  if escolha <= (indice - 1): #o indice da lista começa com 0, por isso preciso subtrair 1
    poluente = opcoes[escolha - 1]
    unidade = unidades[poluente]

    dados_poluente = dados[poluente]

    print(f'IMPRIMINDO MÉDIAS DIÁRIAS PARA O POLUENTE {poluente} \n')

    #https://stackoverflow.com/questions/3294889/iterating-over-dictionaries-using-for-loops
    for data, valores in dados_poluente.items():
      media = sum(valores) / len(valores) #para calcular a média diária
      print(f'Em {data}, a média diária foi {media:.2f} {unidade}')

  elif escolha == 6:
    for poluente in opcoes:
      dados_poluente = dados[poluente]
      unidade = unidades[poluente]

      print(f'IMPRIMINDO MÉDIAS DIÁRIAS PARA O POLUENTE {poluente}')

      for data, valores in dados_poluente.items():
        media = sum(valores) / len(valores)
        print(f'Em {data}, a média diária foi {media:.2f} {unidade}')

      print('\n---------------------------------------\n')

  else:
    print('Essa opção é inválida')
    print('Tente novamente, dentre as opções do índice.')

else:
  print('Essa opção é inválida')
  print('Tente novamente, dentre as opções do índice.')
```
