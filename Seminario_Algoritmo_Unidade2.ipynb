```python
import csv #https://docs.python.org/3/library/csv.html
#importa biblioteca que vai ler o arquivo CSV

import urllib.request #https://docs.python.org/3/library/urllib.request.html#module-urllib.request
#importa biblioteca que vai ler o link do github com o arquivo CSV

url_arquivo_dados = 'https://raw.githubusercontent.com/klaramyk/TrabalhoN1_Tema6_Poluicao_Atmosferica/main/suape_janeiro_2022.csv'
```

```python
with urllib.request.urlopen(url_arquivo_dados) as leitor:
  linhas = leitor.read().decode('utf-8').splitlines()
  dados = {} #dict para armazenar os valores do poluente como data:valor
  unidades = {} #dict para armazenar a unidade de medida geral

  #itera sobre o arquivo para definir a posição dos dados necessários
  for linha in csv.reader(linhas[1:]):
    data = linha[0]
    poluente = linha[4]
    valor = float(linha[5])
    unidade = linha[6]




    unidades[poluente] = unidade

    if poluente in dados:

      if data in dados[poluente]:
        dados[poluente][data].append(valor)
        #adiciona o valor no final da lista já existente

      else:
        dados[poluente][data] = [valor]
        #adiciona valor na lista de poluente, caso não exista ainda

    else:
      dados[poluente] = {data: [valor]}
      #adiciona novo poluente = chave:valor na lista

opcoes = list(dados.keys())
#exibe todas as chaves do dict (nomes dos poluentes) como um índice
#https://stackoverflow.com/questions/16819222/how-do-i-return-dictionary-keys-as-a-list-in-python
```
